```python
import requests

def av_request (function, symbol, apikey = ALPHA_VANTAGE_API_KEY):
    url = f'https://www.alphavantage.co/query?function={function}&symbol={symbol}&apikey={apikey}'
    r = requests.get(url)
    return r.json()
```

## Overview

```python
def overview(symbol):
    return av_request('OVERVIEW', symbol)

overview(SYMBOL)
```

## Income

```python
def income_statement(symbol):
    return av_request('INCOME_STATEMENT', symbol)

income_statement(SYMBOL)
```

## Balance Sheet

```python
def balance_sheet(symbol):
    return av_request('BALANCE_SHEET', symbol)

balance_sheet(SYMBOL)
```

## Cash Flows

```python
def cash_flow(symbol):
    return av_request('CASH_FLOW', symbol)

cash_flow(SYMBOL)
```
