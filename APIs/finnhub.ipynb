```python
import requests

def fh_req(function, symbol, api_key=API_KEY):
    url = f'https://finnhub.io/api/v1/stock/{function}?symbol={symbol}&token={api_key}'
    response = requests.get(url)
    return response.json()
```

```python
def financials_reported(symbol):
    return fh_req('financials-reported', symbol)

financials_reported(SYMBOL)
```

```python
def recommendation(symbol):
    return fh_req('recommendation', symbol)

recommendation(SYMBOL)
```

```python
def insider_sentiment(symbol):
    return fh_req('insider-sentiment', symbol)

insider_sentiment(SYMBOL)
```
