```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_FILE = './data/us_data.csv'
```

```python
df = pd.read_csv(DATA_FILE, index_col='ticker')
df.describe()
```

```python
col='quickRatio'
x1 = df[df['zacks_rank'] == 1.0][[col]]
x2 = df[df['zacks_rank'] == 2.0][[col]]
x3 = df[df['zacks_rank'] == 3.0][[col]]
x4 = df[df['zacks_rank'] == 4.0][[col]]
x5 = df[df['zacks_rank'] == 5.0][[col]]
```

```python
kwargs = dict(alpha=0.5, bins=5)

plt.hist(x5, **kwargs, color='blue', label='#5')
plt.hist(x4, **kwargs, color='red', label='#4')
#plt.hist(x3, **kwargs, color='green', label='#3')
plt.hist(x2, **kwargs, color='yellow', label='#2')
plt.hist(x1, **kwargs, color='black', label='#1')

#plt.gca().set(title='Frequency Histogram of Diamond Depths', ylabel='Frequency')
#plt.xlim(50,75)
plt.legend()
```
