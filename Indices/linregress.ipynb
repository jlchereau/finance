```python
import numpy as np
import pandas as pd
import yfinance as yf
from scipy.stats import linregress
```

## Growth

```python
ticker = yf.Ticker('AAPL')
#data = ticker.actions
#data = ticker.dividends
data = ticker.financials
#data = ticker.history(period='5y')
#data = ticker.info
data
```

```python
x = (data.columns.year - data.columns[-1:].year).to_numpy()
y = data.loc['Total Revenue',:].to_numpy().astype('float64')
result = linregress(x, y)
print(result.slope / y[0])
```

```python
x = (data.columns.year - data.columns[-1:].year).to_numpy()
y = data.loc['Normalized EBITDA',:].to_numpy().astype('float64')
result = linregress(x, y)
print(result.slope / y[0])
```

```python
x = (data.columns.year - data.columns[-1:].year).to_numpy()
y = data.loc['Normalized Income',:].to_numpy().astype('float64')
result = linregress(x, y)
print(result.slope / y[0])
```

```python
x = (data.columns.year - data.columns[-1:].year).to_numpy()
y = data.loc['Research And Development',:].to_numpy().astype('float64')
result = linregress(x, y)
print(result.slope / y[0])
```
