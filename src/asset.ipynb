```python
import numpy as np
df = equity.yahoo_ticker.history(period='5y', interval='1d', auto_adjust=True)
df['Return'] = df['Close'].astype(float).pct_change()
df.head(10)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, height_ratios=(2/3, 1/3))

# Plot the price line charts
ax1.plot(df.index, df['Close'], color='blue', label='Close Price')


ax2.bar(df.index, df['Volume'], color='red', alpha=0.3, label='Volume')


plt.show()
```
